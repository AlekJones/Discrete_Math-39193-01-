## 1 Graphs and Trees

For my graph i'll be using Magic the Gathering formats as examples and the flow will be where to start in you collection building based on those formats.

```
import graphviz

dot = graphviz.Digraph(comment='Formats')
dot.node('A', 'Limited') #Don't have to actually own any cards, drafting cards from booster packs.
dot.node('B', 'Standard') #The most recent rotation of cards
dot.node('C', 'Pioneer') #Basically standard plus recent sets
dot.node('D', 'Modern') #The shift from casual to competitive
dot.node('E', 'Legacy') #Vintage-lite
dot.node('F', 'Vintage') #The pinnacle format
dot.node('G', 'Commander') #Can be played with a little or a lot

dot
```

The basic formats of the TCG Magic the Gathering.

```
dot.edges(['AB', 'AG']) #Limited generally leads to standard and opens commander
dot.edges(['BC', 'AC']) #Standard rotates into Pioneer
dot.edges(['AD','CD'])
dot.edges(['AE','DE'])
dot.edges(['AF','EF'])
dot
```

Essentially, as you can see you could theoretically play every format from just limited.

I'm not sure why I have so many different sets of lines from Limited but the degree should be 6. You can start with limited and branch off in developing your collection from there. Enough playing Limited you could play any format!
